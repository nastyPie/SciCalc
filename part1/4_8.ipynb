Написать класс для вычисления определенного функции двух переменных в единичном
квадрате методом Монте-Карло. Для этого сгенерировать на единичном квадрате n пар
случайных чисел (xi, yi), вычислить fi= f(xi, yi,, дополнительные_параметры), вычислить их
сумму значений функции. Предусмотреть передачу дополнительных параметров.
Исследовать погрешность в зависимости от n на примере простой функции.

```python
import random
import math
import matplotlib.pyplot as plt


class Integ:
    def __init__(self, func, x1, x2, y1, y2, n, *args): #3
        self.func = func #4
        self.x1 = x1 #5
        self.x2 = x2 #6
        self.y1 = y1 #7
        self.y2 = y2 #8
        self.n = n #9
        self.args = args #10

    
    def __str__(self): #12
        return f'Функция: {self.func.__name__}\nКол-во точек: {self.n}\nПределы интегрирования: x[{self.x1},{self.x2}] \
y[{self.y1},{self.y2}]' #13

    
    def __repr__(self):
        return self.__str__()

    
    def gen_n(self, n): #21
        x = [random.uniform(self.x1, self.x2) for _ in range(n)] #22
        y = [random.uniform(self.y1, self.y2) for _ in range(n)] #23
        return list(zip(x, y)) #24

    
    def calc(self, n=0): #17
        if n == 0: #18
            n = self.n #19
        L = self.gen_n(n) #20
        s = sum([self.func(*i, *self.args) for i in L]) #25
        return (s/n)*((self.x2-self.x1)*(self.y2-self.y1)) #26

    
    def draw(self): #28
        n_count = range(1000) #29
        y = [self(i) for i in n_count] #30
        plt.plot(n_count, y) #31
        plt.show() #32


    def __call__(self, n): #15
        return self.calc(n) #16


def func(x, y):
    return math.sin(x)*math.cos(y)

def func2(x, y, a):
    return x*a

a = 100 #1
obj = Integ(func2, 1, 4, 1, 4, 1000, a) #2
print(obj) #11
print(obj(10000)) #14
obj.draw() #27
obj(100000) #33
```
