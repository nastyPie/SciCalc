–†–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º —Å–ª—É—á–∞–π–Ω–æ–µ –±–ª—É–∂–¥–∞–Ω–∏–µ. –ü—É—Å—Ç—å —á–∞—Å—Ç–∏—Ü—ã –≤ –Ω–∞—á–∞–ª—å–Ω—ã–π –º–æ–º–µ–Ω—Ç –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ —Ç–æ—á–∫–µ
—Å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ (x=0, y=0). –ù–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ —á–∞—Å—Ç–∏—Ü–∞ –ø–æ–ª—É—á–∞–µ—Ç —Å–ª—É—á–∞–π–Ω—ã–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ –Ω–∞ [-1. 1] –ø—Ä–∏—Ä–∞—â–µ–Ω–∏—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç x, y. –ü—Ä–æ–º–æ–¥–µ–ª–∏—Ä–æ–≤–∞—Ç—å 10000 —à–∞–≥–æ–≤,
–Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å —Å–ª—É—á–∞–π–Ω—ã–µ –±–ª—É–∂–¥–∞–Ω–∏—è —á–∞—Å—Ç–∏—Ü—ã –Ω–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏. –ö—Ä–æ–º–µ —ç—Ç–æ–≥–æ, –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫, –∫–∞–∫
–º–µ–Ω—è–µ—Ç—Å—è —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ —á–∞—Å—Ç–∏—Ü –æ—Ç –Ω–∞—á–∞–ª–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç (‚àöùë•^2 + ‚àöy^2)^0.5 –æ—Ç –Ω–æ–º–µ—Ä–∞ —à–∞–≥–∞. 

```python
import random
import matplotlib.pyplot as plt


class Particle:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        self.list_of_movement = [(self.x, self.y)]
        self.distance_to_0_0 = [(0, 0)]

    def make_a_step(self):
        dx = random.uniform(-1, 1)
        dy = random.uniform(-1, 1)
        self.x += dx
        self.y += dy

    def random_walk(self, n):
        for i in range(1, n+1):
            self.make_a_step()
            self.list_of_movement.append((self.x, self.y))
            self.distance_to_0_0.append((((self.x**2 + self.y**2)**.5), i))

    def draw(self):
        xs = [i[0] for i in self.list_of_movement]
        ys = [i[1] for i in self.list_of_movement]
        xd = [i[1] for i in self.distance_to_0_0]
        yd = [i[0] for i in self.distance_to_0_0]

        fig = plt.figure(figsize=(10, 10))
        ax1 = fig.add_subplot(1, 2, 1)
        ax1.set(title='–ü–æ–ª–æ–∂–µ–Ω–∏–µ —á–∞—Å—Ç–∏—Ü—ã')
        ax1.set_xlabel('x')
        ax1.set_ylabel('y')
        ax2 = fig.add_subplot(1, 2, 2)
        ax2.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ 0, 0 –æ—Ç –Ω–æ–º–µ—Ä–∞ —à–∞–≥–∞')
        ax2.set_xlabel('–ù–æ–º–µ—Ä —à–∞–≥–∞')
        ax2.set_ylabel('–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ –Ω–∞—á–∞–ª–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç')
        ax1.plot(xs, ys)
        ax2.plot(xd, yd)
        plt.tight_layout()
        plt.show()


part = Particle()
part.random_walk(10000)
part.draw()
```

```python
class Particle2(Particle):
    def make_a_step(self):
        dx = random.gauss(0, 1)
        dy = random.gauss(0, 1)
        self.x += dx
        self.y += dy


part2 = Particle2()
part2.random_walk(10000)
part2.draw()
```
