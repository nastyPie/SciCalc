1. Создать генераторное выражение, содержащее 1000 нечетных чисел. Превратить генераторное выражение в список.
2. С помощью matplotlib визуализировать получившийся список, К оси ординат примените логарифмическую шкалу. 

```python
%matplotlib inline
import matplotlib.pyplot as plt


#Генераторное выражение содержащие 1000 нечётных чисел
G = (i for i in range(1, 2000) if i%2 != 0)
#Превращение генераторного выражения в список
list_G = list(G)

x = [i for i in range(1, 1001)]

plt.plot(x, list_G)
#К оси ординат применяем логарифмическую шкалу
plt.yscale('log')
plt.grid()
plt.show()
```
