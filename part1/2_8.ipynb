```python
import matplotlib.pyplot as plt
import numpy as np



def foo(n):
	return {i:i**2 for i in range(1, n+1)}


def draw(n):
	fig, ax = plt.subplots()
	D = foo(n)
	x = np.array([key for key in D])
	y = np.array([D[key] for key in D]) 
	ax.plot(x, y)
	plt.xlabel('Ключи')
	plt.ylabel('Значения')
	plt.show()


draw(200)
```
