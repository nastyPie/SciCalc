Написать класс для вычисления глобальных максимума и минимума функции f(x,
дополнительные_параметры) на произвольном отрезке [a, b] методом Монте-Карло. Для
этого на отрезке [a, b] создается n равномерно распределенных случайных чисел,
вычисляются для них значения функции, берется минимум и максимум. Класс должен
обеспечить вычислением глобального минимума и максимума для любой функции, любого
числа случайных чисел и на любом отрезке.

```python
import numpy as np


def f(x, *args):
    return -10*x**2 + 1.01*x + 3.2


def f2(x, *args):
    return np.sin(x)/x


class MaxMinFinder:
    def __init__(self, a, b):
        self.a = a
        self.b = b


    def find(self, func, n, *args):
        x = [np.random.uniform(self.a, self.b) for _ in range(n)]
        y = [func(xx, *args) for xx in x]
        return min(y), max(y)
    
    def __call__(self, func, n, *args):
        return self.find(func, n, *args)


finder = MaxMinFinder(-0.6, 0.6)
tup = finder.find(f, 1000)
print(tup)
```

```python
finder2 = MaxMinFinder(0, 15)
tup2 = finder2.find(f2, 1000)
print(tup2)
```
