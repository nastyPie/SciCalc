2. Ğ¡ Ğ¿Ğ¾Ğ¼Ğ¾Ñ‰ÑŒÑ Ğ»ÑĞ±Ğ¾Ğ¹ Ğ¿Ñ€Ğ¾Ñ†ĞµĞ´ÑƒÑ€Ñ‹ Ğ²Ñ‹Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ¸Ñ ĞºĞ¾Ñ€Ğ½ĞµĞ¹ Ğ¿Ğ¾ÑÑ‚Ñ€Ğ¾Ğ¹Ñ‚Ğµ Ğ³Ñ€Ğ°Ñ„Ğ¸Ğº Ğ·Ğ°Ğ²Ğ¸ÑĞ¸Ğ¼Ğ¾ÑÑ‚Ğ¸ Ğ¾Ğ´Ğ½Ğ¾Ğ³Ğ¾ Ğ¸Ğ· ĞºĞ¾Ñ€Ğ½ĞµĞ¹ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ğ¸ Ğ¿.1 Ğ¾Ñ‚ 0.2<=a<=2

```python
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import numpy as np


def func(x, a):
	return 1/np.tan(a*x) - x**3


def roots():
	root_list = [fsolve(func, .1, args=a)[0] for a in (.1, .2, 1, 2)]
	return root_list


def draw():
	fig = plt.figure(figsize=(12,7))
	ax_1 = fig.add_subplot(2, 3, 1)
	ax_2 = fig.add_subplot(2, 3, 2)
	ax_3 = fig.add_subplot(2, 3, 4)
	ax_4 = fig.add_subplot(2, 3, 5)
	ax_1.set(title='a=0.1')
	ax_2.set(title='a=0.2')
	ax_3.set(title='a=1')
	ax_4.set(title='a=2')

	ax_5 = fig.add_subplot(2, 3, 3)
	ax_5.set(title='Ğ—Ğ°Ğ²Ğ¸ÑĞ¸Ğ¼Ğ¾ÑÑ‚ÑŒ ĞºĞ¾Ñ€Ğ½Ñ Ğ¾Ñ‚ Ğ° (Ğ¿Ñ€Ğ¸Ğ±Ğ»Ğ¸Ğ¶ĞµĞ½Ğ¸Ğµ 0.1)')

	x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
	x0 = [0 for i in range(len(x))]
	for ax, a in zip((ax_1, ax_2, ax_3, ax_4), (.1, .2, 1, 2)):
		ax.plot(x, func(x, a), x, x0)
		ax.set_ylim([-50, 50])
	ax_5.plot((.1, .2, 1, 2), roots())
	plt.show()


draw()
```
