#### Условие задачи:
1. Написать функцию, которой передаются параметры n, m, s. С помощью спискового включения сгенерировать n случайных   нормально распределённых чисел с математическим ожиданием m и средним квадратичным отклонением s (см. библиотеку random). Удерживать в списке только положительные случайные числа.
2. Вызвать функцию для n=1000, m=1, s=0.5. С помощью matplotlib нарисовать получившуюся в первой задаче последовательность случайных чисел.

```python
import random
import matplotlib.pyplot as plt


def func1(m, s):
    a = random.normalvariate(m, s)
    while a <= 0:
        a = random.normalvariate(m, s)
    return a


def func2(n, m, s):
	lst = [func1(m, s) for i in range(n)]
	return lst


def draw(n, m, s):
	fig, ax = plt.subplots() # Получаем оси и пространство для фигуры (Создаем Figure и помещаем на нее Axes)
	y = func2(n, m, s) # Преобразовываем список сгенерированных чисел в массив
	x = [i for i in range(1, len(y)+1)] # Преобразовываем список номеров этих чисел в массив
	fig.set_figheight(10)
	fig.set_figwidth(10)
	ax.plot(x, y) # Рисуем точки, соединенные линиями
	plt.xlabel('Номер числа')
	plt.ylabel('Значение')
	plt.show()


draw(1000, 1, 0.5)
```
