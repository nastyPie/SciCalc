Написать класс для работы с матрицами размером n*n. Класс должен создавать матрицы,
обеспечивать их аккуратную распечатку, сложение и вычитание, проверять корректность
переданных конструктору данных. Должны выполняться следующие действия
Matrix([[1,2],[3,4]]) - Matrix([[5,6],[7,8]])


```python
import numpy as np


class Matrix:
    def __init__(self, L):
        for i in L:
            if not isinstance(i, list):
                raise ValueError('Wrong input')
            for j in i:
                if type(j) not in (int, float):
                    raise ValueError('Wrong input')
        self.matrix = np.array(L)

    def __str__(self):
        return str(self.matrix)

    def __repr__(self):
        return self.__str__()

    def __add__(self, other):
        if isinstance(other, Matrix):
            try:
                return Matrix((self.matrix+other.matrix).tolist())
            except ValueError:
                raise ValueError('Matrices must be of the same dimension')

    def __sub__(self, other):
        if isinstance(other, Matrix):
            try:
                return Matrix((self.matrix-other.matrix).tolist())
            except ValueError:
                raise ValueError('Matrices must be of the same dimension')


ans = Matrix([[1,2],[3,4]]) - Matrix([[5,6],[7,8]])
m1 = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print(ans)
m1 + Matrix([[101,102],[103, 104]])
```

```python
class Matrix2:
    def __init__(self, L):
        for i in L:
            if not isinstance(i, list):
                raise ValueError('Wrong input')
            for j in i:
                if type(j) not in (int, float):
                    raise ValueError('Wrong input')
        self.matrix = L

    def __str__(self):
        return '['+'\n '.join(map(str, self.matrix))+']'

    def __repr__(self):
        return self.__str__()

    def __add__(self, other):
        if isinstance(other, Matrix2):
            res = []
            for i in range(len(self.matrix)):
                try:
                    res.append([self.matrix[i][j] + other.matrix[i][j] for j in range(len(self.matrix))])
                except IndexError:
                    raise ValueError('Matrices must have the form n * n and / or be of the same dimension')
            return Matrix2(res)

    def __sub__(self, other):
        if isinstance(other, Matrix2):
            res = []
            for i in range(len(self.matrix)):
                try:
                    res.append([self.matrix[i][j] - other.matrix[i][j] for j in range(len(self.matrix))])
                except IndexError:
                    raise ValueError('Matrices must have the form n * n and / or be of the same dimension')
            return Matrix2(res)


m1 = Matrix2([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
m2 = Matrix2([[1, 2], [3, 4]])
m3 = Matrix2([[10, 2000], [30, 40]])
m4 = m2 + m3
m5 = m2 - m3
print(m2)
print(m3)
print(m4)
print(m5)
m1 + Matrix2([[101,102],[103, 104]])
```
