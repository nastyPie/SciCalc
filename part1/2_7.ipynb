1. Создать пустой кортеж. Кортеж с единственным элементом. Напишите функцию, которой передается параметр n. Функция возвращает кортеж из n элементов, элементами являются квадраты индексов элементов в кортеже.
2. Вызвать функцию для n = 100. С помощью matplotlib нарисовать график для возвращаемого кортежа из первой задачи. Отобразите на рисунке надписи на осях, сетку, заголовок.

```python
%matplotlib inline
import matplotlib.pyplot as plt


tup1 = () # Пустой кортеж
tup2 = (0,) # Кортеж с единственным элементом


def func1(n):
	return tuple([i**2 for i in range(n)])



y = func1(100) #Список значений функции
x = [i[0] for i in enumerate(y)] #Список индексов значений функции

#Рисуем график
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Степенная функция')
plt.grid(linestyle='solid', linewidth=0.3)
plt.show()
```
