Написать класс для игры в рулетку. Вы должны обеспечить случайное выкидывание одного
из чисел от 0 до 36, кроме того, должны определить является ли результат четным (красное)
или нечетным (черное). Особым образом обработать выпадение 0 (зеро). Должны быть
методы, обеспечивающие отображение результата в виде строки (два метода).


```python
import random


class Roulette:

    def roll(self):
        num = random.randint(0, 36)
        if num == 0:
            self.result = (0, 'Zero')
        elif num % 2 == 0:
            self.result = num, 'Красное'
        else:
            self.result = num, 'Чёрное'
        return self.result

    def __str__(self):
        try:
            return f'{self.result}'
        except AttributeError:
            print('Roll the roulette!')
            return ''

    def __repr__(self):
        return self.__str__()


game = Roulette()
game.roll()
print(game)
game.roll()
print(game)
game.roll()
print(game)
game.roll()
print(game)
game.roll()
print(game)
game.roll()
print(game)
```
