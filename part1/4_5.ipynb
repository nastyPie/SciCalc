Написать класс для генерации случайных чисел, распределённых нормально и равномерно.
Результатом должен быть список случайных чисел заданной длины, распределенный по
заданном закону. Закон задается при создании экземпляра класса. Получить его можно с
помощью свойства random_data.

```python
import random


class Random_number:
    def __init__(self, a, b, n):
        self.a = a
        self.b = b
        self.n = n

    def random_data(self, dist, n):
        if dist == 'uniform':
            return [random.uniform(self.a, self.b) for _ in range(n)]
        elif dist == 'normal':
            return [random.gauss(self.a, self.b) for _ in range(n)]

    def __call__(self, dist, n):
        return self.random_data(dist, n)


rd = Random_number(1, 2, 10000)
rd('uniform', 100)
```
