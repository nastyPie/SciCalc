Создать класс для работы с полиномами. Класс должен создавать полином у(x) =
a0+a1*x+a2*x^2+...+an*x^n, заданный его коэффициентами, вычислять значение полинома для
произвольного x, вычислять суммы и разности двух полиномов одинаковой степени.

```python
import numpy as np


class Polinom:
    def __init__(self, *coef):
        self.pol = np.poly1d(coef[::-1])

    def calculate(self, x):
        return self.pol(x)
    
    def __call__(self, x):
        return self.calculate(x)

    def __str__(self):
        return str(self.pol)
    
    def __repr__(self):
        return self.__str__()

    def __add__(self, other):
        if isinstance(other, Polinom):
            coef = tuple((self.pol + other.pol).c)
            return Polinom(*coef[::-1])
        if isinstance(other, (int, float)):
            return Polinom(*((self.pol + other).c)[::-1])
        raise ValueError

    def __radd__(self, other):
        return self.__add__(other)

    def __sub__(self, other):
        if isinstance(other, Polinom):
            coef = tuple((self.pol - other.pol).c)
            return Polinom(*coef[::-1])
        if isinstance(other, (int, float)):
            return Polinom(*((self.pol - other).c)[::-1])
        raise ValueError

    def __rsub__(self, other):
        return self.__sub__(other)



pol1 = Polinom(2, 3, 4)
pol2 = Polinom(5, 6, 7)
print(pol1.calculate(10), pol2.calculate(8))
print(pol1-pol2)
print(pol1+pol2)
```
