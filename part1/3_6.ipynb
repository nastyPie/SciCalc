1. Исследуйте положение первого нуля функции Ганкеля Hq(x) от порядка q. Функция вычисляется так:

    import scipy

    scipy.special.hankel1(q, x).real

    здесь x – последовательность значений, для которых вычисляется функция, 0<=q<1 – порядок.

    Постройте графики hankel1 (q,x), где 0<=x<=10, q=0, 0.1, 0.2,0.3, , …. 0.9.,1.0

2. С помощью любой процедуры решения уравнений вычислите и нарисуйте зависимость положения первого нуля функции от значения  модуля q=0, 0.01, 0.02, 0.03, …0.98, 0.99. 1.0

```python
import matplotlib.pyplot as plt
import scipy.special as sc
from scipy.optimize import fsolve
from numpy import linspace


def hankel(x, q):
    return sc.hankel1(q, x).real


def func(q):
    x = [i for i in linspace(0, 10, 201)]
    y = [hankel(xx, q) for xx in x]
    return x, y


def root_finder():
    return [fsolve(hankel, 3, args=q)[0] for q in list(linspace(0, 1, 101))]


def draw():
    fig = plt.figure(figsize=(15,10))
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.set_xlabel('x')
    ax1.set_ylabel('hankel1(q, x)')
    ax1.set_title('функция Ганкеля')
    ax2 = fig.add_subplot(1, 2, 2)
    for q in list(linspace(0, 1, 11)):
        x, y = func(q)
        ax1.plot(x, y, label=q)
    ax1.plot(x, [0 for i in x])
    ax1.grid()
    ax1.legend()
    
    roots = root_finder()
    ax2.plot(list(linspace(0, 1, 101)), roots)
    ax2.set_xlabel('q')
    ax2.set_ylabel('root')
    ax2.set_title('Зависимость корня от q')
    plt.show()


draw()
```
