Написать класс Mammal для представления млекопитающего. У класса имеются атрибуты
mammal_kind – вид животного и имя – name. Напишите методы для обеспечения
представления млекопитающего.
Написать два класса Dog – собака и Cat – кот, наследующие от Mammal, Продумайте, что
нужно изменить в этих классах.

```python
from random import choice

class Mammal:
    def __init__(self, mammal_kind, name):
        self.mammal_kind = mammal_kind
        self.name = name

    def __str__(self):
        return f'{self.mammal_kind} - {self.name}'

    def walk(self):
        print(f'{self.name} is walking')

    def breathe(self):
        print(f'{self.name} is breathing')

    def voice(self):
        print(f'{self.mammal_kind} {self.name} makes some kind of sound')


class Dog(Mammal):
    def __init__(self, name):
        super().__init__('Dog', name)
        self.name = name

    def voice(self):
        print(f'{self.name} says WOOF')


class Cat(Mammal):
    def __init__(self, sex, name=None):
        super().__init__('Cat', name)
        if sex in ('Male', 'Female'):
            self.sex = sex
        else: raise ValueError(f'Wrong gender: {sex}')
    
    def __repr__(self):
        return f'Cat: sex-{self.sex}, name-{self.name}'

    def voice(self):
        print(f'{self.name} says MEOW')
    
        
    def set_name(self, name):
        if self.name != None:
            print(f'The cat already has a name: {self.name}')
        else:
            self.name = name

    def __add__(self, other):
        if isinstance(other, Cat):
            if self.sex == 'Male':
                if other.sex == 'Female':
                    return Cat(sex=choice(('Male', 'Female')))
                raise ValueError('Same-sex reproduction is impossible')
            elif self.sex == 'Female':
                if other.sex == 'Male':
                    return Cat(sex=choice(('Male', 'Female')))
                raise ValueError('Same-sex reproduction is impossible')

cat1 = Cat('Male', 'Barsik')
print(cat1)
cat2 = Cat('Female', 'Alice')
print(cat2)
cat4 = Cat('Male', 'Pushok')
print(cat1 + cat2)
try:
    print(cat1 + cat4)
except ValueError as er:
    print(er)
```
