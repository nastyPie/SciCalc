1.  Написать функцию, которой передаются n, a, b. В функции сгенерировать данные для
    рисования фигуры Лиссажу t=[0,1/n, 2/n, 3/n,… 1], x = cos(aπt), y=sin(bπt). Функция
    возвращает списки t, x, y.
2.  Вызвать функцию для n= 1000, a = 2.6, b=3.7. С помощью matplotlib нарисовать
    фигуру Лиссажу, снабдив рисунок заголовком, сеткой, надписями на осях.

```python
import matplotlib.pyplot as plt
from math import sin, cos, pi


def lissajous(n, a, b):
	t = [i/n for i in range(n+1)]
	x = [cos(a*pi*i) for i in t]
	y = [sin(b*pi*i) for i in t]
	return t, x, y

t, x, y = lissajous(1000, 2.6, 3.7)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Фигура Лиссажу')
plt.grid(linestyle='solid', linewidth=0.3)
plt.show()
```

### Доп. задача
Напишите функцию, которая бы использовала написанную Вами функцию и осуществляла рисование фигуры.
В этой функции:
Введите параметр для толщины линии
Введите параметр для размеров рисунка
Сделайте несколько вызовов для рисования фигур при различных a, b

```python
def lissajous2(n, a, b, figsize=(4,4), lw=3):
    '''
    n- число разбиений
    a, b- коэффициенты
    figsize- размер рисунка
    lw- толщина линии
    '''
    t, x, y = lissajous(n, a, b)
    plt.figure(figsize=figsize)
    plt.plot(x, y, linewidth=lw)
    plt.show()

lissajous2(1000, 3.1, 7.9)
lissajous2(20, 5.1, 6.2, lw=5)
lissajous2(2000, 1.3, 9.1, figsize=(6,6))
```
