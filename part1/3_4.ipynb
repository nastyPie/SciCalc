## –ö–ú3


1) –†–∞—Å—Å–º–æ—Ç—Ä–∏—Ç–µ —ç–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫—É—é —Å—Ö–µ–º—É, —Å–æ—Å—Ç–æ—è—â—É—é –∏–∑ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –≠–î–° (E), —Ä–µ–∑–∏—Å—Ç–æ—Ä–∞ –∏ –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ NE, –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –≤–æ–ª—å—Ç-–∞–º–ø–µ—Ä–Ω–æ–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–æ–π (–í–ê–•): ùëñùëÅùê∏=ùë¢ùëÅùê∏^3, –≥–¥–µ ùëñùëÅùê∏, ùë¢ùëÅùê∏ ‚Äì —Ç–æ–∫ –∏ –Ω–∞–ø—Ä—è–∂–µ–Ω–∏–µ –Ω–∞ –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–º —ç–ª–µ–º–µ–Ω—Ç–µ., ùê∏(ùë°)=cos‚Å°(ùë°). –°–æ—Å—Ç–∞–≤—å—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ü–µ–ø–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ ùë¢ùëÅùê∏.

2) –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –Ω–∞–ø—Ä—è–∂–µ–Ω–∏—è –∏ —Ç–æ–∫–∞ –Ω–∞ –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–º —ç–ª–µ–º–µ–Ω—Ç–µ –ø—Ä–∏ R=1 –û–º –∏ R=0.2 –û–º. –î–ª—è —ç—Ç–æ–≥–æ –í–∞–º –ø—Ä–∏–¥–µ—Ç—Å—è —Ä–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ E(t). –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–µ–π ùëñùëÅùê∏(ùë°), ùë¢ùëÅùê∏(ùë°) –Ω–∞ –¥–≤—É—Ö —Ä–∏—Å—É–Ω–∫–∞—Ö –æ—Ç –≤—Ä–µ–º–µ–Ω–∏ –æ—Ç 0 –¥–æ 10.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve


def vah(u):
	return u**3

def schem(Une, r, e):
	return Une + vah(Une)*r - e

def E(t):
	return np.cos(t)

def draw():
	R1 = 1
	R2 = 0.2
	x = [t for t in np.linspace(0, 10, 1000)]
	#–†–µ—à–∞–µ–º scheme –¥–ª—è –∫–∞–∂–¥–æ–≥–æ t –ø—Ä–∏ R
	y1_u = [fsolve(schem, 0, args=(R1, E(t)))[0] for t in x] 
	y1_i = [vah(u) for u in y1_u]
	y2_u = [fsolve(schem, 0, args=(R2, E(t)))[0] for t in x]
	y2_i = [vah(u) for u in y2_u]


	fig = plt.figure()
	ax1 = fig.add_subplot(2, 2, 1)
	ax1.set(title='U –Ω–∞ NE –ø—Ä–∏ R=1')
	ax1.set_xlabel('–í—Ä–µ–º—è')
	ax1.set_ylabel('U –Ω–∞ NE')
	ax2 = fig.add_subplot(2, 2, 2)
	ax2.set(title='I –Ω–∞ NE –ø—Ä–∏ R=1')
	ax2.set_xlabel('–í—Ä–µ–º—è')
	ax2.set_ylabel('–¢–æ–∫ –Ω–∞ NE')
	ax3 = fig.add_subplot(2, 2, 3)
	ax3.set(title='U –Ω–∞ NE –ø—Ä–∏ R=0.2')
	ax3.set_xlabel('–í—Ä–µ–º—è')
	ax3.set_ylabel('U –Ω–∞ NE')
	ax4 = fig.add_subplot(2, 2, 4)
	ax4.set(title='I –Ω–∞ NE –ø—Ä–∏ R=0.2')
	ax4.set_xlabel('–í—Ä–µ–º—è')
	ax4.set_ylabel('–¢–æ–∫ –Ω–∞ NE')
	ax1.plot(x, y1_u)
	ax2.plot(x, y1_i)
	ax3.plot(x, y2_u)
	ax4.plot(x, y2_i)
	plt.tight_layout()
	plt.show()


draw()
```

### –ö–ú 3 
–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–æ–π –ø–µ—Ä–µ–¥–∞—é—Ç—Å—è E- –≠–î–° –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –∏ R - —Å–æ–ø—Ä–æ—Ç–∏–≤–ª–µ–Ω–∏–µ —Ä–µ–∑–∏—Å—Ç–æ—Ä–∞.

–§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤—ã–¥–∞–≤–∞—Ç—å –¥–≤–∞ —Ä–∏—Å—É–Ω–∫–∞:

–Ω–∞ –ø–µ—Ä–≤–æ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –Ω–∞–ø—Ä—è–∂–µ–Ω–∏—è –Ω–∞ —Ä–µ–∑–∏—Å—Ç–æ—Ä–µ –∏ –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–º —ç–ª–µ–º–µ–Ω—Ç–µ –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –≠–î–° –æ—Ç -E –¥–æ +E.
–Ω–∞ –≤—Ç–æ—Ä–æ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Ç–æ–∫–∞ –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –≠–î–° –æ—Ç -E –¥–æ +E.
–í —É—Å–ª–æ–≤–∏—è—Ö –∑–∞–¥–∞—á–∏, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, —Å–¥–µ–ª–∞–π—Ç–µ —Ç–∞–∫, —á—Ç–æ–±—ã –æ—Ç–æ–±—Ä–∞–∂–∞–ª–∏—Å—å –Ω–∞–¥—Å—Ç—Ä–æ—á–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã.

```python
def func(E, R):
    x = [i for i in np.linspace(-E, E, 1000)]
    Ur = [i for i in x]
    Une = [i for i in x]
    Ir = [i/R for i in x]
    Ine = [vah(i) for i in x]


    fig = plt.figure()
    ax1 = fig.add_subplot(2, 2, 1)
    ax1.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Ur –æ—Ç E')
    ax1.set_xlabel('E')
    ax1.set_ylabel('Ur')
    ax2 = fig.add_subplot(2, 2, 2)
    ax2.set_xlabel('E')
    ax2.set_ylabel('Une')
    ax2.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Une –æ—Ç E')
    ax3 = fig.add_subplot(2, 2, 3)
    ax3.set_xlabel('E')
    ax3.set_ylabel('Ir')
    ax3.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Ir –æ—Ç E')
    ax4 = fig.add_subplot(2, 2, 4)
    ax4.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Ine –æ—Ç E')
    ax4.set_xlabel('E')
    ax4.set_ylabel('Ine')
    ax1.plot(x, Ur)
    ax2.plot(x, Une)
    ax3.plot(x, Ir)
    ax4.plot(x, Ine)
    plt.tight_layout()
    plt.show()

func(10, 10)
```

```python
def draw2(E, R):
    x = [i for i in np.linspace(-E, E, 1000)]
    Une = [fsolve(schem, 0, args=(R, e))[0] for e in x]
    Ur = [E-une for une in Une]
    Ine = [vah(une) for une in Une]
    Ir = [ur/R for ur in Ur]

    fig = plt.figure()
    ax1 = fig.add_subplot(2, 2, 1)
    ax1.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Ur –æ—Ç E')
    ax1.set_xlabel('E')
    ax1.set_ylabel('Ur')
    ax2 = fig.add_subplot(2, 2, 2)
    ax2.set_xlabel('E')
    ax2.set_ylabel('Une')
    ax2.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Une –æ—Ç E')
    ax3 = fig.add_subplot(2, 2, 3)
    ax3.set_xlabel('E')
    ax3.set_ylabel('Ir')
    ax3.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Ir –æ—Ç E')
    ax4 = fig.add_subplot(2, 2, 4)
    ax4.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Ine –æ—Ç E')
    ax4.set_xlabel('E')
    ax4.set_ylabel('Ine')
    ax1.plot(x, Ur)
    ax2.plot(x, Une)
    ax3.plot(x, Ir)
    ax4.plot(x, Ine)
    plt.tight_layout()
    plt.show()

draw2(10, 10)
```
