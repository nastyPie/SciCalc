1) –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ –∫–æ—Ä–Ω–µ–π —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —É—Ä–∞–≤–Ω–µ–Ω–∏—è cos(ùë•)‚àíùë°ùëî(ùëé‚àôùë•)=0 –¥–ª—è 0‚â§ùë•‚â§1.5 –∏ 0.1‚â§a‚â§1.

2) –° –ø–æ–º–æ—â—å—é –ª—é–±–æ–π –ø—Ä–æ—Ü–µ–¥—É—Ä—ã —Ä–µ—à–µ–Ω–∏—è —É—Ä–∞–≤–Ω–µ–Ω–∏–π –≤—ã—á–∏—Å–ª–∏—Ç–µ –∏ –Ω–∞—Ä–∏—Å—É–π—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∫–æ—Ä–Ω—è —É—Ä–∞–≤–Ω–µ–Ω–∏—è –æ—Ç a= 0.01, 0.02, 0.03, ‚Ä¶0.97, 0.98, 0.99, 1.0

```python
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import numpy as np


def func(x, a):
    return np.cos(x) - np.tan(a*x)


def main():
    fig = plt.figure(figsize=(15,10))
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.set_xlabel('x')
    ax1.set_ylabel('func(x)')
    ax1.set(title='–ü–æ–ª–æ–∂–µ–Ω–∏–µ –∫–æ—Ä–Ω–µ–π –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö –∞')
    ax2 = fig.add_subplot(1, 2, 2)
    ax2.set_xlabel('a')
    ax2.set_ylabel('–ö–æ—Ä–µ–Ω—å –ø—Ä–∏ –∞ –∏–∑ [0.01, 1]')
    ax2.set(title='–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∫–æ—Ä–Ω—è –æ—Ç –∞')

    x = [i for i in np.linspace(0, 1.5, 200)]
    for a in [aa for aa in np.linspace(0.1, 1, 4)]:
        y = [func(xx, a) for xx in x]
        ax1.plot(x, y, label=round(a, 1))
    ax1.plot(x, [0 for i in range(len(x))])
    ax1.legend(title='a')
    ax1.grid()

    ax1.set_ylim((-5, 2))
    a = np.linspace(0.01, 1, 100)
    root_list = [fsolve(func, 1, args=aa)[0] for aa in a]
    ax2.plot(a, root_list)
    plt.show()


main()
```
