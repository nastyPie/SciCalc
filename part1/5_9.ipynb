```python
import random
import matplotlib.pyplot as plt


class Dice:
    def roll(self):
        return random.randint(1, 6)


class Game:
    def __init__(self, dice_count, throws):
        self.dice_count = dice_count
        self.throws = throws
        self.dices = [Dice() for dice in range(dice_count)]

    def play(self):
        scores = {dice_dots: 0 for dice_dots in range(1*len(self.dices), 6*len(self.dices) + 1)}
        self.rolls = dict()
        for i in range(1, self.throws+1):
            s = sum(map(lambda x: x.roll(), self.dices))
            self.rolls[i] = s
            scores[s] += 1
        self.res = [scores[dice_dots] for dice_dots in range(1*len(self.dices), 6*len(self.dices) + 1)]
        return self.res

    def view(self):
        index = list(range(len(1*self.dices), 6*len(self.dices) + 1))
        plt.bar(index, self.res)
        plt.xlabel('Выпавшее число')
        plt.ylabel('Кол-во раз')
        plt.show()

    def freq_calc(self, *args):
        s = sum(self.rolls.values())
        freqs = dict()
        for f in args:
            freqs[f] = (list(self.rolls.values()).count(f))/self.throws
        print(f'Сумма всех выпавших очков: {s}')
        print(f'Частоты выпадения {args}:', freqs)

    def __call__(self, dice_count, throws):
        self.__init__(dice_count, throws)
        return self.play()

    def __str__(self):
        res = ''
        for key in self.rolls:
            res += f'Бросок {key}: {self.rolls[key]}\n'
        return res[:-1]

    def __repr__(self):
        return self.__str__()

game = Game(5, 10000)
game(5, 10000)
game.freq_calc(6, 7, 29, 30)
game.view()
```
