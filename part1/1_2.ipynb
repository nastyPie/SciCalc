```python
import turtle


def top():
	turtle.forward(40)
	turtle.left(150)
	turtle.forward(90)
	turtle.left(60)
	turtle.forward(90)
	turtle.left(150)
	turtle.forward(40)


def bottom():
	turtle.right(90)
	turtle.forward(50)
	turtle.left(90)
	turtle.forward(75)
	turtle.left(90)
	turtle.forward(50)



def arrow():
	turtle.reset()
	turtle.pensize(7)
	turtle.color('#003366')
	turtle.down()
	turtle.begin_fill()
	top()
	bottom()
	turtle.fillcolor('#3366FF')
	turtle.end_fill()


arrow()
turtle.mainloop()
```

```python
def fib(n):
	if n < 2:
		return n
	return fib(n - 1) + fib(n - 2)


def test(n):
	'''
		Тестовая функция, выводящая первые n чисел фиббоначи
	'''
	for i in range(1, n+1):
		print(fib(i), end=' ')


test(15)
```
