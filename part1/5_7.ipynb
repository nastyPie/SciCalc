Написать класс для игры в кости. Бросается не кубик, а тетраэдр. Возможное число
выкидываемых очков: 1, 2, 3. 4. Обеспечить случайное бросание кости (случайным образом
выкидывается от одного до четырех очков). 10000 раз случайным образом кидаются десять
костей. Постройте гистограмму распределения очков, каждый раз выбрасывается от 10 до 40
очков. Предусмотрите это при построении гистограммы.


```python
import random
import matplotlib.pyplot as plt


class Dice:
    def roll(self):
        return random.randint(1, 4)


class Game:
    def __init__(self, dice_count, throws):
        self.dice_count = dice_count
        self.throws = throws
        self.dices = [Dice() for dice in range(dice_count)]
        self.points = range(self.dice_count, 4*self.dice_count+1)
        self.res = [0 for _ in range(len(self.points))]
        self.rolls = dict()

    def play(self):
        scores = {points: 0 for points in self.points}
        for i in range(1, self.throws+1):
            s = sum([dice.roll() for dice in self.dices])
            scores[s] += 1
            self.rolls[i] = s
        self.res = [scores[key] for key in scores]
        return self.res

    def view(self):
        plt.bar(self.points, self.res)
        plt.xlabel('Выпавшее число')
        plt.ylabel('Кол-во раз')
        plt.show()

    def __call__(self, dice_count=10, throws=10000):
        self.__init__(dice_count, throws)
        return self.play()

    def __str__(self):
        res = ''
        for key in self.rolls:
            res += f'Бросок {key}: {self.rolls[key]}\n'
        return res[:-1]
    
    def __repr__(self):
        return self.__str__()


game = Game(10, 10000)
game()
game.view()
```
