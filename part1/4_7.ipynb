Кватернион обобщение комплексных чисел. Q = a +b*i + c*j + d*k, т.е. кватернион задается
четверкой чисел a, b, c, d. Конструктор должен работать при задании любого числа
параметров, например, если задано Q(1,2), то создается кватернион 1 +2*i + 0*j + 0*k.
Создать методы для представления кватернионов, их сложения и вычитания.


```python
class Q:
    def __init__(self, a=0, b=0, c=0 ,d=0):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
        self.L = [a, b, c, d]

    def __str__(self):
        res = f'{self.a}+{self.b}*i+{self.c}*j+{self.d}*k'
        return res.replace('+-', '-')

    def __repr__(self):
        return self.__str__()

    def __add__(self, other):
        if isinstance(other, Q):
            a = self.a + other.a
            b = self.b + other.b
            c = self.c + other.c
            d = self.d + other.d
            return Q(a, b, c, d)
        elif isinstance(other, (int, float)):
            a = self.a + other
            return Q(a, self.b, self.c, self.d)
        raise ValueError(f'{other} must be a Quatterion')

    def __sub__(self, other):
        if isinstance(other, Q):
            a = self.a - other.a
            b = self.b - other.b
            c = self.c - other.c
            d = self.d - other.d
            return Q(a, b, c, d)
        elif isinstance(other, (int, float)):
            a = self.a - other
            return Q(a, self.b, self.c, self.d)
        raise ValueError(f'{other} must be a Quatterion')

    def __rsub__(self, other):
        return self.__sub__(other)



q1 = Q(1, 2, -3, 4)
q2 = Q(10, 100)
q3 = Q(0 , 0, 0, 0)
print(q1+q2)
print(q1-q2)
print(q3)
```
