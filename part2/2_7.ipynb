Нарисуйте график этой функции на отрезке [-2, 2].

2) Напишите функцию, которая генерирует матрицу размером n*m,
n>m? элементами которой являются равномерно распределенные на
отрезке [1-1]

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x, T=1):
    t = x % T
    try:
        y = np.zeros(x.shape[0])
    except (IndexError, AttributeError):
        y = np.zeros(1)
    y = np.where((0 <= t) & (t < 1/3), t, y)
    y = np.where((1/3 <= t) & (t < 2/3), 1, y)
    y = np.where((2/3 <= t) & (t < 1), -1, y)
    return y

def view1(a, b, n):
    x = np.linspace(a, b, n)
    y = func(x)
    plt.plot(x, y)
    plt.show()

def view2(a, b, n):
    x = np.linspace(a, b, n)
    y = [func(i) for i in x]
    plt.plot(x, y)
    plt.show()

view1(-2, 2, 1000)
```

```python
view2(-2, 2, 10000)
```

```python
def gen_matrix(n, m, a=-1, b=1):
    if n > m:
        return np.random.uniform(a, b, (n, m))
    else:
        raise AttributeError('n > m!')

try:
    print(gen_matrix(2, 10))
except AttributeError:
    print('Выброшено исключение')
print(gen_matrix(6, 5))
```
