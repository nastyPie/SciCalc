2) Из изображения пункта 1 сделайте цветное изображение. Красная составляющая совпадает с исходным изображением, зеленая составляющая с отображенным в зеркале, все элементы синей составляющее равны 255. Циклы не использовать. Выведите получившееся изображение.

```python
from scipy.misc import ascent
import matplotlib.pyplot as plt


img = ascent()
print(img.shape) #Размерность

print(type(img[0]))
print(type(img[0, 0])) #Тип элементов массива

print(img.nbytes) #Объём занимаемой памяти

plt.imshow(img, cmap='gray')
```

```python
img2 = img[::-1]
plt.imshow(img2, cmap='gray') #Отображение относительно осью абсцисс
```
