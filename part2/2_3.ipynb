
1) Написать функцию, которой на вход передаётся массив, возвращает функция среднее значение элементов массива, медиану, среднее квадратичное отклонение.<br>
2) Сгенерировать массив 200x200 случайных равномерно распределенных на [0,1] чисел. Сделать это можно это можно с помощью numpy.random.uniform() (см. документацию). Построить график результатов применения функции, написанной в п.1. от k=2, 3,…200. Функции должны передаваться квадратные срезы исходного массива размерности k*k.

```python
import numpy as np
import matplotlib.pyplot as plt


def get_calc(arr):
    median = np.median(arr)
    mean = np.mean(arr)
    std = np.std(arr)
    return median, mean, std


def gen_matrix(m, n, a, b):
    return np.random.uniform(a, b, (m, n))


def draw(matrix, k):
    result = np.zeros((len(k), 3))
    for i in range(len(k)):
        result[i] = get_calc(matrix[:k[i], :k[i]])
    fig = plt.figure(figsize=(15, 5))
    ax1 = fig.add_subplot(131)
    ax2 = fig.add_subplot(132)
    ax3 = fig.add_subplot(133)
    ax1.set_title('Медиана от k')
    ax2.set_title('Среднее значение от k')
    ax3.set_title('Среднее квадратичное отклонение от k')
    ax1.plot(k, result[:, 0])
    ax2.plot(k, result[:, 1])
    ax3.plot(k, result[:, 2])
    plt.tight_layout()
    plt.show()
    

matrix = gen_matrix(200, 200, 0, 1)
k = range(2, 201)
draw(matrix, k)
```
