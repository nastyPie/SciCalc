Сконструировать массив x0= np.linspace(0, 1, n). С помощью NumPy выполнять преобразования $x_{i+1}=r*x_i*(1-x_i)$, i=0,1,…m, при этом массив $x_{i+1}(t)$ изменяетcя. Анимировать, как этот массив будет менять от i. В пользовательский интерфейс вынести: n, 0<=r<4, m, а также флажок, по которому предыдущие графики будут стираться или не стираться. Запуск анимации осуществляется по нажатию кнопки. Пользовательский интерфейс сверстать в два столбца. Приложение должно работать в Voila.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
from ipywidgets import widgets
from time import sleep


def change_arr(x, r):
    res = r*x*(1-x)
    return res
    

def draw_arr(x, i, ax):
    ax.set_title(f'Изменение массива x, i={i}')
    ax.set_xlabel('x')
    ax.set_xlim((-10, len(x)+10))
    ax.set_ylim((-0.1, 1))
    ax.plot(range(len(x)), x)


def main(*args):
    n = w_n.value
    r = w_r.value
    m = w_m.value
    delay = w_delay.value
    x = np.linspace(0, 1, n)
    list_x = []
    with out2:
        for i in range(m):
            clear_output(wait=True)
            fig = plt.figure()
            ax = fig.add_subplot(111)
            if w_clear.value:
                draw_arr(x, i, ax)
            else:
                list_x.append(x)
                for xx in list_x:
                    draw_arr(xx, i, ax)
            x = change_arr(x, r)
            sleep(delay)
            plt.show()


out1 = widgets.Output(layout={'width': '50%'})
out2 = widgets.Output(layout={'width': '50%'})
n_label = widgets.Label('Число точек n')
w_n = widgets.IntSlider(description='', min=0, max=1000, step=1, value=500)
r_label = widgets.Label('Коэффициент r')
w_r = widgets.FloatSlider(description='', min=0, max=4, step=0.01, value=1)
m_label = widgets.Label('Число кадров m')
w_m = widgets.IntSlider(description='', min=0, max=100, step=1, value=10)
clear_label = widgets.Label('Очистка предыдущих графиков')
w_clear = widgets.RadioButtons(options={'Да': True,
                                       'Нет': False,})
delay_label = widgets.Label('Задержка между кадрами')
w_delay = widgets.FloatSlider(description='', min=0, max=1, step=0.01, value=0.1)
button = widgets.Button(description='Запустить', button_style='primary')
button.on_click(main)

display(widgets.HBox([out1, out2]))
with out1:
    display(n_label, w_n,
            r_label, w_r,
            m_label, w_m,
            clear_label, w_clear,
            delay_label, w_delay,
            button)
```
