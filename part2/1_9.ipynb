1) Из библиотеки scipy.misc импортируйте функцию ascent. Результат вызова этой функции представляет собой изображение в градациях серого. Нанести на это изображения черную сетку через 30 пикселей (черное соответствуют 0). Циклы не применять, отобразить поучившееся.

2) Поучившееся изображение удвоить, добавив к изображению п.1 еще одно такое же изображение снизу от первого.

```python
#Задание 1
from scipy.misc import ascent
import matplotlib.pyplot as plt
import numpy as np


asc = ascent()
asc[::30] = 0
asc[:, ::30] = 0
plt.imshow(asc, cmap='gray')
```

```python
#Задание 2
a = np.append(asc, asc, axis=0)
plt.imshow(a, cmap='gray')
```
