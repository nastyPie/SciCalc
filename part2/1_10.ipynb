### Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import face, ascent


f, a = face(gray=True), ascent()
print(f'Размер face: {f.shape}\nРазмер ascent: {a.shape}')
print(f'Максимум из ширин изображений: {max(f.shape[1], a.shape[1])}')
print(f'Максимум из высот изображений: {max(f.shape[0], a.shape[0])}')


def merge(A, B, r, c):
    A[r:r+B.shape[0], c:c+B.shape[1]] = B
    return A


def func(img1, img2):
    res = np.zeros((img1.shape[0], img1.shape[1]))
    res = merge(res, img2, 0, 0)
    res = merge(res, img2, 0, img2.shape[1])
    res = merge(res, img2[0:img1.shape[0]-img2.shape[0]], img2.shape[1], 0)
    res = merge(res, img2[0:img1.shape[0]-img2.shape[0]], img2.shape[1], img2.shape[1])
    return res

img = func(f, a)
plt.imshow(img, cmap='gray')
```
