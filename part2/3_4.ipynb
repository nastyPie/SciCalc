## Заданиие 1

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
from ipywidgets import widgets
from time import sleep
from scipy.integrate import quad


def get_integ(a, b, n, func):
    ax = np.linspace(a, b, n)
    i = np.array(range(n))
    h = abs((b - a)/n)
    res = np.zeros(n)
    res[i] = func(ax[i] + h/2)
    return h*np.sum(res)
    
    
def main(*args):
    a = w_a.value
    x = w_x.value
    n = n_partions.value
    func = function.value
    xx = np.linspace(w_a.value, w_x.value, n_partions.value)
    yy = func(xx) 
    L1 = np.zeros(n)
    L2 = np.zeros(n)
    for i in range(n-1):
        aa, bb = xx[i], xx[i+1]
        standart = quad(func, aa, bb)[0]
        integ = get_integ(aa, bb, n, func)
        L1[i+1] = L1[i] + standart
        L2[i] = (integ - standart) / standart
    with out2:
        clear_output(wait=True)
        fig = plt.figure(figsize=(5, 10))
        ax1 = fig.add_subplot(3, 1, 1)
        ax1.plot(xx, yy)
        ax1.set_title('График функции')
        ax2 = fig.add_subplot(3, 1, 2)
        ax2.plot(xx, L1)
        ax2.set_title('График интеграла')
        ax3 = fig.add_subplot(3, 1, 3)
        ax3.plot(xx, L2)
        ax3.set_title('Относительаня погрешность')
        plt.show()


out1 = widgets.Output(layout={'width': '50%'})
out2 = widgets.Output(layout={'width': '50%'})
function_label = widgets.Label('Выбор функции')
function = widgets.RadioButtons(options={'f(x) = x': lambda x: x,
                                         'f(x) = x^2': lambda x: x**2,
                                         'f(x) = sin(x)': lambda x: np.sin(x),
                                         'f(x) = exp(x)': lambda x: np.exp(x)})
section_label = widgets.Label('Границы интегрирования')
w_a = widgets.FloatSlider(description='a', min=-100, max=100, step=0.01, value=0)
w_x = widgets.FloatSlider(description='x', min=-100, max=100, step=0.1, value=10)
n_partions_label = widgets.Label('Число разбиений отрезка')
n_partions = widgets.IntSlider(description='', min=0, max=1000, step=1, value=101)
button = widgets.Button(description='Запустить', button_style='primary')
button.on_click(main)
display(widgets.HBox([out1, out2]))
with out1:
    display(function_label, function,
           section_label, w_a, w_x,
           n_partions_label, n_partions,
           button)
```
