На квадрате n*n на верхней стороне сгенерировать случайное распределение частиц (от 1 до n-1). Построить анимацию распространения частиц вниз. Частица в следующем слое появляется, если в предыдущем слое с ней смежны от k1 до k2 частиц. В пользовательский интерфейс выносятся: число случайно расположенных частиц в верхнем слое от 1 до n-1, а также n, m, k1, k2, интервал между кадрами. Анимируйте распространение частиц от верхнего к нижнему слою. Вывод можно сделать с помощью функции matplotlib imshow().

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
from ipywidgets import widgets
from time import sleep


def gen_rand_lvl(n, count):
    res = np.zeros(n)
    i = np.array(range(count))
    res[i] = 1
    np.random.shuffle(res)
    return res

def get_next_lvl(pr_lvl, k1, k2):
    new_lvl = np.zeros(len(pr_lvl))
    pr_lvl = np.insert(np.zeros(2), 1, pr_lvl)
    for i in range(1, len(pr_lvl)-1):
        if k1 <= (pr_lvl[i-1], pr_lvl[i], pr_lvl[i+1]).count(1) <= k2:
            new_lvl[i-1] = 1
    return new_lvl

#tests
k1, k2 = 1, 2
n, count = 12, 6
pr_lvl = gen_rand_lvl(n, count)
next_lvl1 = get_next_lvl(pr_lvl, k1, k2)
next_lvl2 = get_next_lvl(next_lvl1, k1, k2)
pr_lvl, next_lvl1, next_lvl2 
```
