```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets
from IPython.display import display, clear_output


def H(x, eps, a):
    y = np.zeros(x.shape[0])
    y = np.where(x < -eps, 0, y)
    y = np.where((-eps <= x) & (x <= eps), a/2+a*x/(2*eps)+a/(2*np.pi)*np.sin(np.pi*x/eps), y)
    y = np.where(x > eps, a, y)
    return y
```

```python
def func(b): 
    N = w_g.value
    xmin = w_c.value 
    xmax = w_d.value
    eps = w_a.value
    a = w_b.value
    x = np.linspace(xmin, xmax, N) 
    y = H(x, eps, a)
    
    with w_out2: 
        clear_output(wait=True)
        plt.figure(figsize=(4,4)) 
        plt.plot(x, y, 'k-', lw=2)
        plt.xlabel('x')
        plt.ylabel(r'$H(x,\epsilon,a)$')
        plt.grid(True)
        plt.title('График функции') 
        plt.show() 

# Верстка интерфейса         
w_title = widgets.HTML('<h4>Исследование функции</h4>') 

w_a_L = widgets.Label('eps') 
w_a = widgets.FloatSlider(description='', min=-5., max=5., 
                          step=0.01, value=1.)
w_b_L = widgets.Label('a')
w_b = widgets.FloatSlider(description='', min=-10., max=10., 
                          step=0.01, value=1)
w_c_L = widgets.Label('xmin') 
w_c = widgets.FloatSlider(description='', min=-10., max=10., 
                          step=0.01, value=-10.)
w_d_L = widgets.Label('xmax')
w_d = widgets.FloatSlider(description='', min=-10., max=10., 
                          step=0.01, value=10)
w_g_L = widgets.Label('N')
w_g = widgets.IntSlider(description='', min=2, max=100, 
                          step=1, value=50)

w_btn = widgets.Button(description="Запустить", button_style='info') 
w_btn.on_click(func) 
w_out2 = widgets.Output() 

w_gui=widgets.HBox([widgets.VBox([w_title, widgets.HBox([w_a_L, w_a]),
                    widgets.HBox([w_b_L, w_b]),
                    widgets.HBox([w_c_L, w_c]),
                    widgets.HBox([w_d_L, w_d]),
                    widgets.HBox([w_g_L, w_g]),
                    w_btn]), 
                    w_out2]) 
display(w_gui) 
```
