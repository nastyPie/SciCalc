1) Загрузите енота в градациях серого. Переставьсте четные и нечетные строки и столбцы изображения. Для этого вам понадобится копия изображения, которую можно создать с помощью numpy.copy(). Отобразите результат. Циклы неиспользовать.<br>
2) Увеличьте изображения п1, продублировав сначала каждую строку и каждый столбец. Отобразите получившееся изображение

```python
from scipy.misc import face
import matplotlib.pyplot as plt
import numpy as np


f = face(gray=True)

f_tmp = np.copy(f)
f_tmp[::2], f_tmp[1::2] = f[1::2], f[::2]
f_copy = np.copy(f_tmp)
f_copy[::, ::2], f_copy[::, 1::2] = f_tmp[::, 1::2], f_tmp[::, ::2]
print(f[:6], '\n\n')
print(f_copy[:6])

plt.imshow(f_copy, cmap='gray')
```

## Задание 2

```python
def increase_img(img, n):
    tmp = np.repeat(img, n, axis=0)
    return np.repeat(tmp, n, axis=1)


new_img = increase_img(f_copy, 2)
print(new_img[:6], '\n\n')
print(f_copy[:6])
plt.imshow(new_img, cmap='gray')
```
