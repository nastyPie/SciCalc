Имя палитры передается в именованном параметре cmap. Посмотреть, какие имеются палитры, можно, задав произвольное имя палитры.

```python
from scipy.misc import face
import matplotlib.pyplot as plt


f = face(gray=True)
plt.imshow(f, cmap='gray')
```

```python
f_n = 255 - f
plt.imshow(f_n, cmap='gray')
```

```python
def decrease(img, step):
    return img[::step, ::step]

f_n_d = decrease(f_n, 5)
plt.imshow(f_n_d, cmap='Spectral')
```

```python
plt.imshow(f_n_d, cmap='Paired')
```

```python
plt.imshow(f_n_d, cmap='jet')
```
