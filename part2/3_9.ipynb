## КМ 3

1. В начальный момент времени в единичном квадрате n частиц расположены в
точке (0.5, 0.5). На каждом шаге моделирования частицы получают случайные
приращения dx=np.random.uniform(-s, s), dy=np.random.uniform(-s, s) или
dx=np.random.normal(0, s), dy=np.random.normal(0, s). Постройте
пользовательский интерфейс для моделирования случайных блужданий. В
пользовательском интерфейсе необходимо задавать число частиц, параметр s,
размеры области, число шагов моделирования, закон распределения. При
моделировании отображается конечное распределение частиц, средние и
средние значения координат и средние квадратичные отклонения координат
частиц от времени, радиус окружности, описанной вокруг частиц.

2. Анимируйте случайные блуждания частиц в единичном квадрате. В заголовок
графика анимации вынесите шаг анимации и число частиц, находящихся в
единичном квадрате в данный момент. В пользовательский интерфейс
вынесите n, s, число шагов анимации, задержка между шагами.

```python
from ipywidgets import widgets
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, clear_output


def get_calcs(x_arr, y_arr):
    avg_x = np.sum(x_arr)/len(x_arr)
    avg_y = np.sum(y_arr)/len(y_arr)
    msd_x = np.sum((x_arr - avg_x)**2)/len(x_arr)
    msd_y = np.sum((y_arr - avg_y)**2)/len(y_arr)
    radius = max(max(x_arr) - min(x_arr), max(y_arr) - min(y_arr))/2
    median_x = np.median(x_arr)
    median_y = np.median(y_arr)
    return msd_x, msd_y, avg_x, avg_y, radius, median_x, median_y

def main(*args):
    part_count = w_n.value
    s = w_s.value
    a = w_a.value
    b = w_b.value
    step_count = w_step_count.value
    distribution = w_dist.value
    x_arr = np.zeros(part_count) + 0.5
    y_arr = np.zeros(part_count) + 0.5
    calcs_list = np.zeros((step_count, 7))
    for step in range(step_count):
        for i in range(len(x_arr)):
            if distribution == 0:
                x_arr[i] += np.random.uniform(-s, s)
                y_arr[i] += np.random.uniform(-s, s)
            elif distribution == 1:
                x_arr[i] += np.random.uniform(0, s)
                y_arr[i] += np.random.uniform(0, s)
        calcs_list[step] = get_calcs(x_arr, y_arr)
    with out2:
        clear_output(wait=True)
        fig = plt.figure(figsize=(10, 10))
        ax1 = fig.add_subplot(4, 2, 1)
        ax1.set_xlim(-a, a)
        ax1.set_ylim(-b, b)
        ax1.set_title('Конечное расположение частиц')
        ax1.plot(x_arr, y_arr, 'ro')
        ax2 = fig.add_subplot(4, 2, 2)
        ax2.set_title('Среднеквадратичное отклонение по X от времени')
        ax3 = fig.add_subplot(4, 2, 3)
        ax3.set_title('Среднеквадратичное отклонение по Y от времени')
        ax2.plot(range(step_count), calcs_list[::, 0])
        ax3.plot(range(step_count), calcs_list[::, 1])
        ax4 = fig.add_subplot(4, 2, 4)
        ax5 = fig.add_subplot(4, 2, 5)
        ax4.set_title('Среднее значение X от времени')
        ax5.set_title('Среднее значение Y от времени')
        ax4.plot(range(step_count), calcs_list[::, 2])
        ax5.plot(range(step_count), calcs_list[::, 3])
        ax6 = fig.add_subplot(4, 2, 6)
        ax6.set_title('Радиус описанной окружности от времени')
        ax6.plot(range(step_count), calcs_list[::, 4])
        ax7 = fig.add_subplot(4, 2, 7)
        ax7.set_title('Медиана X от времени')
        ax8 = fig.add_subplot(4, 2, 8)
        ax8.set_title('Медиана Y от времени')
        ax7.plot(range(step_count), calcs_list[::, 5])
        ax8.plot(range(step_count), calcs_list[::, 6])
        plt.tight_layout()
        plt.show()


out1 = widgets.Output(layout={'width': '50%'})
out2 = widgets.Output(layout={'width': '50%'})
w_title = widgets.HTML('<h3>Моделирование случайного блуждания частиц</h3>')
w_n_L = widgets.Label('Колличество частиц')
w_n = widgets.IntSlider(description='', min=0, max=100, step=1, value=10)
w_s_L = widgets.Label('S')
w_s = widgets.FloatSlider(description='', min=0, max=1, step=0.01, value=0.1)
w_a_L = widgets.Label('Размер области (а)')
w_a = widgets.IntSlider(description='', min=1, max=100, step=1, value=3)
w_b_L = widgets.Label('Размер области (b)')
w_b = widgets.IntSlider(description='', min=1, max=100, step=1, value=3)
w_step_count_L = widgets.Label('Колличество шагов')
w_step_count = widgets.IntSlider(description='', min=0, max=1000, step=1, value=100)
w_dist_L = widgets.Label('Распределение')
w_dist = widgets.RadioButtons(options={'-s, s':0, '0, s':1})
w_button = widgets.Button(description='Запустить', button_style='primary')
w_button.on_click(main)
display(widgets.HBox([out1, out2]))
with out1:
    display(w_title,
            w_n_L, w_n,
            w_s_L, w_s,
            w_a_L, w_a,
            w_b_L, w_b,
            w_step_count_L, w_step_count,
            w_dist_L, w_dist,
            w_button)
```

## Задание 2
Анимируйте случайные блуждания частиц в единичном квадрате. В заголовок графика анимации вынесите шаг анимации и число частиц, находящихся в единичном квадрате в данный момент. В пользовательский интерфейс вынесите n, s, число шагов анимации, задержка между шагами.

```python
from time import sleep


def sq_check(x_arr, y_arr):
    count = 0
    for i in range(len(x_arr)):
        if 0 <= x_arr[i] <= 1 and 0 <= y_arr[i] <= 1:
            count += 1
    return count


def run(*args):
    n = w_n.value
    s = w_s.value
    step = w_step.value
    delay = w_delay.value
    x_arr = np.zeros(n) + 0.5
    y_arr = np.zeros(n) + 0.5 
    with out2:
        for i in range(step):
            for j in range(len(x_arr)):
                x_arr[j] += np.random.uniform(-s, s)
                y_arr[j] += np.random.uniform(-s, s)
            count = sq_check(x_arr, y_arr)
            clear_output(wait=True)
            plt.plot(x_arr, y_arr, 'ro')
            plt.xlim(0, 1)
            plt.ylim(0, 1)
            plt.title(f'Шаг:{i}, Число частиц в квадрате: {count}')
            plt.show()
            sleep(delay)
        

out1 = widgets.Output(layout={'width': '50%'})
out2 = widgets.Output(layout={'width': '50%'})
n_L = widgets.Label('Колличество частиц')
w_n = widgets.IntSlider(description='', min=0, max=100, step=1, value=10)
s_L = widgets.Label('S')
w_s = widgets.FloatSlider(description='', min=0, max=1, step=0.01, value=0.1)
step_L = widgets.Label('Колличество шагов анимации')
w_step = widgets.IntSlider(description='', min=10, max=200, step=1, value=100)
delay_L = widgets.Label('Задержка между шагами')
w_delay = widgets.FloatSlider(description='', min=0, max=1, step=0.01, value=0.1)
button = widgets.Button(description='Запустить', button_style='primary')
button.on_click(run)
display(widgets.HBox([out1, out2]))
with out1:
    display(n_L, w_n,
           s_L, w_s,
           step_L, w_step,
           delay_L, w_delay,
           button)
```
