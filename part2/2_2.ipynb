Постройте черный квадрат Малевича: массив n*n, все элементы
которого на расстоянии int(n/5) от краев равны 255 (белый цвет), а
остальные равны нулю. Нарисовать черный квадрат средствами
matplotlib. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
def malevich(n):
    a = np.zeros((n, n))
    a.fill(255)
    a[int(n/5):n-int(n/5), int(n/5):n-int(n/5)] = 0
    return a
```

```python
img1 = malevich(10)
plt.imshow(a, cmap='gray')
```

```python
img2 = malevich(1000)
plt.imshow(img2, cmap='gray')
```

Для построенного черного квадрата и фиксированного n=100
произведите сингулярное разложение исходной матрицы изображения и
восстановите матрицу, удерживая половину и одну десятую
сингулярных чисел, нарисуйте результаты восстановления.

```python
def svd_compress(image, portion=1.):
    U, sigma, VT = np.linalg.svd(image)
    ns = int(len(sigma)*portion)
    compressed_image = U[:,:ns] @ np.diag(sigma[:ns]) @ VT[:ns,:]
    plt.imshow(compressed_image, cmap='gray')
    plt.title(f'Удержано сингулярных чисел {ns} из {len(sigma)}')
    plt.axis('off')
    
    
n = 100
square = malevich(n)
svd_compress(square, 0.5)
```
