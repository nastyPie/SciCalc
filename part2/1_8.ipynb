2) Изображение из пункта 1 отобразить зеркально относительно горизонтальной и вертикальной осей. Циклы не применять, отобразить поучившееся

```python
from scipy.misc import ascent
import matplotlib.pyplot as plt
import numpy as np


img1 = ascent()
img2 = 255 - img1
img3 = img1[:, ::-1]
res_img = np.ndarray(shape=(img1.shape[0], img1.shape[1], 3), dtype=np.int64)
res_img[:, :, 0] = img3[:, :]
res_img[:, :, 1] = img2[:, :]
res_img[:, :, 2] = img1[:, :]


plt.imshow(res_img)
```

```python
res_img[1, :, :]
```

```python
print(img1[1, :][:3]) # Синее
print(img2[1, :][:3]) # Инверт
print(img3[1, :][:3]) # Красное
```
