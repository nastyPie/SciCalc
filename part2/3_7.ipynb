Сглаживание функции, заданной на массиве $x_i$, i=0, 1,…n осуществляется с помощью формулы 
${y^{k+1}_{i}}$ = $\frac1{4}$ * (${y^{k}_{i-1}}$ + 2*${y^{k}_{i}}$ + ${y^{k}_{i+1}}$), где k – номер итерации (сколько раз эта формула применена к массиву). На массиве x c n+1 элементами на отрезке [a,b] задать функции x+noise(0, s), x^2+noise(0,s), sin(x)+noise(0,s). noise – распределенные нормально случайные числа с нулевым математическим ожиданием и средним квадратичным отклонением s. Построить пользовательский интерфейс и анимацию, которая обеспечивает задание отрезка [a, b], функции из набора, указанного выше, n – числа разбиений отрезка и K – числа сглаживаний, интервала между кадрами. Анимация демонстрирует, как меняется массив y(x) в зависимости от числа сглаживаний. Интерфейс построить в два столбца, анимацию запускать по нажатию кнопки. Уровень шума s задать в логарифмическом масштабе.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
from ipywidgets import widgets
from time import sleep


def noise(loc, s):
    n = w_n.value
    return np.random.normal(loc, s, (1, n))

def func_noise1(x, s):
    return (x + noise(0, s))[0]

def func_noise2(x, s):
    return (x**2 + noise(0, s))[0]

def func_noise3(x, s):
    return (np.sin(x) + noise(0, s))[0]


def smooth(L):
    res = []
    res.append(L[0])
    for i in range(1, len(L)-1):
        res.append(0.25*(L[i-1] + 2*L[i] + L[i+1]))
    res.append(L[len(L)-1])
    return res

def smooth_v2(L):
    i = np.array(range(1, len(L)-1))
    L = np.array(L)
    res = np.zeros(len(L))
    res[0] = L[0]
    res[i] = 0.25*(L[i-1] + 2*L[i] + L[i+1])
    res[-1] = L[-1]
    return res


def run(*args):
    a = w_a.value
    b = w_b.value
    if b < a:
        raise ValueError('a < b')
    func = w_func.value
    n = w_n.value
    k = w_k.value
    delay = w_delay.value
    s = w_s.value
    x = np.linspace(a, b, n)
    y = func(x, s)
    with out2:
        for i in range(1, k+1):
            y = smooth(y)
            y_2 = smooth_v2(y)
            clear_output(wait=True)
            plt.plot(x, y, 'ro')
            plt.plot(x, y_2, 'bo')
            plt.title(f'Число сглаживаний: {i}')
            plt.show()
            sleep(delay)

            
out1 = widgets.Output(layout={'width': '50%'})
out2 = widgets.Output(layout={'width': '50%'})
title = widgets.HTML('<h4>Анимация изменения массива y(x) от числа сглаживаний</h4>')
w_a_L = widgets.Label('Граница отрезка a:')
w_a = widgets.FloatSlider(description='', min=-100, max=100, step=1, value=0)
w_b_L = widgets.Label('Граница отрезка b (b > a):')
w_b = widgets.FloatSlider(description='', min=-100, max=100, step=1, value=10)
w_func_L = widgets.Label('Функция из набора:')
w_func = widgets.RadioButtons(options={'x+noise(0, s)': func_noise1,
                                       'x^2+noise(0,s)': func_noise2,
                                       'sin(x)+noise(0,s)': func_noise3})
w_n_L = widgets.Label('Число разбиений отрезка:')
w_n = widgets.IntSlider(description='', min=0, max=1000, step=1, value=100)
w_k_L = widgets.Label('Число сглаживаний')
w_k = widgets.IntSlider(description='', min=0, max=100, step=1, value=40)
w_delay_L = widgets.Label('Интервал между кадрами')
w_delay = widgets.FloatSlider(description='', min=0, max=1, step=0.01, value=0.2)
w_s_L = widgets.Label('Уровень шума:')
w_s = widgets.FloatLogSlider(description='', min=-1, max=0.1, step=0.01, value=0.2)
button = widgets.Button(description='Запустить', button_style='primary')
button.on_click(run)
display(widgets.HBox([out1, out2]))
with out1:
    display(title,
            w_a_L, w_a,
            w_b_L, w_b,
            w_func_L, w_func,
            w_n_L, w_n,
            w_k_L, w_k,
            w_delay_L, w_delay,
            w_s_L, w_s,
            button)
```
