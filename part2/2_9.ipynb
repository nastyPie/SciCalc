—Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é n*n (n –ø–µ—Ä–µ–¥–∞–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–∏ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞),
—ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ –º–∞—Ç—Ä–∏—Ü—ã —è–≤–ª—è—é—Ç—Å—è —Å–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞, —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ
—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ [1,2]. –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å
–º–∞—Ç—Ä–∏—Ü—ã –∏ —á–∏—Å–ª–æ –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç–∏. –í—ã–∑–æ–≤–∏—Ç–µ —Å–∫–æ–Ω—Å—Ç—Ä—É–∏—Ä–æ–≤–∞–Ω–Ω—É—é
—Ñ—É–Ω–∫—Ü–∏—é m =1000 —Ä–∞–∑ –∏ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–µ
–∑–Ω–∞—á–µ–Ω–∏—è –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—è –∏ —á–∏—Å–ª–∞ –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç–∏. –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ NumPy
–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, T=1):
    t = x % T
    y = np.zeros(x.shape[0])
    y = np.where((0 <= t) & (t < 0.25), 2*(t**.5), y)
    y = np.where((0.25 <= t) & (t < 0.5), 1, y)
    y = np.where((0.5 <= t) & (t < 1), 4*t**2, y)
    return y

x = np.linspace(-2, 2, 1000)
plt.plot(x, f(x))
plt.show()
```

```python
#–ü–µ—Ä–∏–æ–¥
x = np.linspace(0, 1, 1000)
plt.plot(x, f(x))
plt.show()
```

```python
#–ù–∞ –±–æ–ª—å—à–µ–º –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ
x = np.linspace(-20, 20, 1000)
plt.plot(x, f(x))
plt.show()
```

```python
def g(n, a=1, b=2):
    m = np.random.uniform(a, b, (n, n))
    det = np.linalg.det(m)
    cond = np.linalg.cond(m)
    return det, cond

print(g(3))
```

```python
def test(m=1000):
    arr_det = np.zeros(m)
    arr_cond = np.zeros(m)
    for i in range(m):
        arr_det[i], arr_cond[i] = g(10)
    print(f'–ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å: {np.min(arr_det)}')
    print(f'–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å: {np.max(arr_det)}')
    print(f'–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç–∏: {np.min(arr_cond)}')
    print(f'–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç–∏: {np.max(arr_cond)}')


test()        
```
