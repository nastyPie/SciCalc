Средствами NumPy выполните численное дифференцирование ${f^`}$(x) = $\frac{f({x_{i+1}}) - f({x_{i-1}})}{2h}$, где h = $\frac{b-a}{n}$. Для функций x, x**2, sin(x), cos(x), exp(x) обеспечить выбор функции, задание отрезка, числа разбиений, вывод на график производных, вчисленных по аналитическим формулам и численно. Предусмотреть зашумление исходных данных нормально распределенным шумом с математическим ожиданием 0 и средним квадратичным отклонением $\sigma$. Кроме того, обеспечить вывод максимальной ошибки диффиренцирования на отрезке. Сверстать пользовательский интерфейс в два столбца.

```python
def run(*args):
    a = w_a.value
    b = w_b.value
    func = w_func.value[0]
    func_dervt = w_func.value[1]
    delay = w_delay.value
    n = w_n.value
    x = np.linspace(a, b, n)
    y1 = func(x)
    y2 = func_dervt(x)
    with out4:
        for i in range(len(x)):
            clear_output(wait=True)
            plt.plot(x[:i], y1[:i], 'ro')
            plt.plot(x[:i], y2[:i], 'bo')
            plt.xlim(np.min(x), np.max(x))
            plt.ylim(np.min(y1), np.max(y1))
            plt.show()
            sleep(delay)

out3 = widgets.Output(layout={'width': '50%'})
out4 = widgets.Output(layout={'width': '50%'})
w_func_L = widgets.Label('Функция')
w_func = widgets.RadioButtons(options={
                                       'x**2': (func2, func2_dervt),
                                       'x': (func1, func1_dervt), 
                                       'sin(x)': (func3, func3_dervt), 
                                       'cos(x)': (func4, func4_dervt), 
                                       'exp(x)': (func5, func5_dervt)})
a_L = widgets.Label('Граница отрезка a')
w_a = widgets.IntSlider(description='', min=-10, max=10, step=1, value=0)
b_L = widgets.Label('Граница отрезка b')
w_b = widgets.IntSlider(description='', min=-10, max=10, step=1, value=3)
n_L = widgets.Label('Число разбиений отрезка')
w_n = widgets.IntSlider(description='', min=1, max=1000, step=1, value=100)
delay_L = widgets.Label('Задержка между кадрами')
w_delay = widgets.FloatSlider(description='', min=0, max=0.1, step=0.001, value=0.005)
button = widgets.Button(description='Запустить', button_style='primary')
button.on_click(run)
display(widgets.HBox([out3, out4]))
with out3:
    display(w_func_L, w_func,
            a_L, w_a,
            b_L, w_b,
            n_L, w_n,
            delay_L, w_delay,
            button)
```
